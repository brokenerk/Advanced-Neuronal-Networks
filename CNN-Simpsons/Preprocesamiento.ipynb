```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from PIL import Image
import urllib.request
```

```python
csv_test = pd.read_csv('2-google-train.csv', usecols=list(np.arange(0,15)))
csv_test2 = csv_test.copy()
csv_test3 = csv_test.copy()
columns = ['url', 'x1', 'x2', 'y1', 'y2']

csv_test.drop(csv_test.columns[list(np.arange(5,15))], axis=1, inplace=True)
csv_test2.drop(csv_test2.columns[list(np.arange(0,5)) + list(np.arange(10,15))], axis=1, inplace=True)
csv_test3.drop(csv_test3.columns[list(np.arange(0,10))], axis=1, inplace=True)

csv_test.columns = columns
csv_test2.columns = columns
csv_test3.columns = columns

test = pd.concat([csv_test, csv_test2, csv_test3])
test.sort_values("url", inplace=True)
test.drop_duplicates(keep='first',inplace=True)
test.reset_index(drop=True, inplace=True)
test
```

```python
test.to_csv('2-google-train-2.csv', index=False)
```
