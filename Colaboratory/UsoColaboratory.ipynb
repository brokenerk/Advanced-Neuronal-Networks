#Uso de comandos Shell

Colaboratory permite el uso de comandos Shell de linux  para navegar entre los directorios y archivos que se encuentran anexos al ambiente 

!ls 
: Lista el contenido de un directorio. Si lo usamos sin argumentos, lista el contenido del directorio actual o corriente. Por más información

%cd

Cambia el directorio corriente (Change Directory) en que nos encontramos posicionados. Si se utiliza sin argumentos, cambia al directorio raíz del homedir del usuario

!pwd

Imprime el directorio de trabajo actual 

!rm < file > 
  
Elimina archivos o directorios.

Los comnados shell se puede usar dentro de python 

In [4]: contents = !ls

In [5]: print(contents)
['myproject.txt']

In [6]: directory = !pwd

In [7]: print(directory)
