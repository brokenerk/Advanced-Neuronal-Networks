## Map and filter

```python
def cuadrado(x):
    return x*2
    
```

```python
seq=[1,2,3,4]
map(cuadrado,seq)
```

```python
list(map(cuadrado,seq))
```

```python
list(map(lambda var:var**2,seq))
```

```python
funlam=lambda var:var*2
```

```python
funlam(2)
```

```python
list(filter(lambda item:item%2==0,seq))
```
